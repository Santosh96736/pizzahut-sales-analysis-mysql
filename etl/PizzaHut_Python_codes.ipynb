```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
user_name = os.getenv('DB_USER')
user_password = os.getenv('DB_PASSWORD')

engine = create_engine(f'mysql+mysqlconnector://{user_name}:{user_password}@localhost/pizzahut')
```

```python
df_pizza_types = pd.read_csv('pizza_types.csv')

print(df_pizza_types.head())
print(df_pizza_types.isnull().sum())
print(df_pizza_types.duplicated().sum())
print(df_pizza_types.dtypes)

df_pizza_types.to_sql(con = engine, name = 'pizza_types', 
                      if_exists='append', method= 'multi',index = False)
```

```python
df_pizzas = pd.read_csv('pizzas.csv')

print(df_pizzas.head())
print(df_pizzas.isnull().sum())
print(df_pizzas.duplicated().sum())
print(df_pizzas.dtypes)

df_pizzas.to_sql(con = engine, name = 'pizzas', 
                 if_exists= 'append', method= 'multi', index = False)
```

```python
df_orders = pd.read_csv('orders.csv', low_memory= False)

print(df_orders.head())
print(df_orders.isnull().sum())
print(df_orders.duplicated().sum())
print(df_orders.dtypes)


df_orders.to_sql(con = engine, name = 'orders', 
                 if_exists='append', method = 'multi', index = False)
```

```python
df_order_details = pd.read_csv('order_details.csv')

print(df_order_details.head())
print(df_order_details.isnull().sum())
print(df_order_details.duplicated().sum())
print(df_order_details.dtypes)

df_order_details.to_sql(con = engine, name = 'order_details', 
                        if_exists= 'append', method = 'multi', index = False)
```
