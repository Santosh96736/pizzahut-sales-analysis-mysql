```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
user_name = os.getenv('DB_USER')
user_password = os.getenv('DB_PASSWORD')

engine = create_engine(f'mysql+mysqlconnector://{user_name}:{user_password}@localhost/pizzahut')
```

```python
def read_file(csv_file):
    return pd.read_csv(csv_file)
```

```python
df_pizza_types = read_file('pizza_types.csv')
```

```python
df_pizzas = read_file('pizzas.csv')
```

```python
df_orders = read_file('orders.csv')
```

```python
df_order_details = read_file('order_details.csv')
```

```python
def inspect_data(df):
    print(df.head())
    print(df.dtypes)
    print(df.columns)
    print(len(df))
    print(df.isnull().sum())
    print(df.duplicated().sum())
```

```python
inspect_data(df_pizza_types)
```

```python
inspect_data(df_pizzas)
```

```python
inspect_data(df_orders)
```

```python
inspect_data(df_order_details)
```

```python
def load_data(df,table_name):
    df.to_sql(name = table_name, con = engine, if_exists = 'append'
              ,method = 'multi', index = False)
```

```python
load_data(df_pizza_types,'pizza_types')
```

```python
load_data(df_pizzas,'pizzas')
```

```python
load_data(df_orders,'orders')
```

```python
load_data(df_order_details,'order_details')
```
